```python
import numpy as np
import cv2
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
albedo = pd.read_csv('Albedo_Map.csv', header = None)
fe = pd.read_csv('LPFe_Map.csv', header = None)
k = pd.read_csv('LPK_Map.csv', header = None)
th = pd.read_csv('LPTh_Map.csv', header = None)
ti = pd.read_csv('LPTi_Map.csv', header = None)
```

```python
plt.figure(figsize =(10,10))
plt.imshow(albedo, cmap='jet')
plt.title("Lunar Albedo")
plt.colorbar(orientation='horizontal', label= 'Th')
```

```python
plt.figure(figsize =(10,10))

plt.subplot(2, 2, 1)
plt.imshow(fe, cmap='jet')
plt.title('Fe map')
plt.colorbar(orientation='horizontal', label= 'Fe')

plt.subplot(2, 2, 2)
plt.imshow(k, cmap='jet')
plt.title('K map')
plt.colorbar(orientation='horizontal', label= 'k')

plt.subplot(2, 2, 3)
plt.imshow(th, cmap='jet')
plt.title('Th map')
plt.colorbar(orientation='horizontal', label= 'Th')

plt.subplot(2, 2, 4)
plt.imshow(ti, cmap='jet')
plt.title('Ti map')
plt.colorbar(orientation='horizontal', label= 'Ti')
```

```python
albedo_leftHalf = albedo.iloc[:, 0:360].values
albedo_rightHalf = albedo.iloc[:, 360:720].values

fe_leftHalf = fe.iloc[:, 0:360].values
fe_rightHalf = fe.iloc[:, 360:720].values

k_leftHalf = k.iloc[:, 0:360].values
k_rightHalf = k.iloc[:, 360:720].values

th_leftHalf = th.iloc[:, 0:360].values
th_rightHalf = th.iloc[:, 360:720].values

ti_leftHalf = ti.iloc[:, 0:360].values
ti_rightHalf = ti.iloc[:, 360:720].values
```

```python
albedo_left = albedo_leftHalf.reshape(-1,1)
fe_left = fe_leftHalf.reshape(-1,1).squeeze()
k_left = k_leftHalf.reshape(-1,1).squeeze()
th_left = th_leftHalf.reshape(-1,1).squeeze()
ti_left = ti_leftHalf.reshape(-1,1).squeeze()
```

```python
albedo_right = albedo_rightHalf.reshape(-1,1)
fe_right = fe_rightHalf.reshape(-1,1).squeeze()
k_right = k_rightHalf.reshape(-1,1).squeeze()
th_right = th_rightHalf.reshape(-1,1).squeeze()
ti_right = ti_rightHalf.reshape(-1,1).squeeze()
```

```python
X_train = np.transpose(np.array([fe_left, k_left, th_left, ti_left]))
X_train = pd.DataFrame(X_train, columns=['Fe', 'K', 'Th', 'Ti'])
X_train
```

```python
y_train = pd.DataFrame(albedo_left, columns=['Pixel'])
y_train
```

```python
X_test = np.transpose(np.array([fe_right, k_right, th_right, ti_right]))
X_test = pd.DataFrame(X_test, columns=['Fe', 'K', 'Th', 'Ti'])
X_test
```

```python
y_test = pd.DataFrame(albedo_right, columns=['Pixel'])
y_test
```

## <font color='blue'> Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
# Model
rfregressor=RandomForestRegressor(n_estimators=100, random_state = 42)

# Train the model
rfregressor.fit(X_train, y_train.values.ravel())
```

```python
# Test the model
y_predRf = rfregressor.predict(X_test)
Rf_mse = mean_squared_error(y_test,y_predRf)
print("Mean Squared Error :", Rf_mse)
y_predRf = y_predRf.reshape(360,360)
```

### <font color='blue'> AdaBoost Regressor

```python
from sklearn.ensemble import AdaBoostRegressor
# Model
adbregressor=AdaBoostRegressor(n_estimators=1000, random_state=42)

# Train the model
adbregressor.fit(X_train, y_train)
```

```python
# Test the model
y_predAdb = adbregressor.predict(X_test)
Adb_mse = mean_squared_error(y_test,y_predAdb)
print("Mean Squared Error :", Adb_mse)
y_predAdb = y_predAdb.reshape(360,360)
```

### <font color='blue'> XGBoost Regressor

```python
from xgboost import XGBRegressor
#Model
xgbregressor = XGBRegressor(n_estimators=100, random_state = 42)

# Train the model
xgbregressor.fit(X_train, y_train.values.ravel())
```

```python
# Test the model
y_predXgb = xgbregressor.predict(X_test)
Xgb_mse = mean_squared_error(y_test,y_predXgb)
print("Mean Squared Error :",Xgb_mse)
y_predXgb = y_predXgb.reshape(360,360)
```

### <font color='blue'> K-Nearest Neighbor Regressor 

```python
from sklearn.neighbors import KNeighborsRegressor
# Model
knnregressor = KNeighborsRegressor(n_neighbors=100)

# Train the model
knnregressor.fit(X_train, y_train.values.ravel())
```

```python
# Test the model
y_predKnn = knnregressor.predict(X_test)
Knn_mse = mean_squared_error(y_test,y_predKnn)
print("Mean Squared Error :",Knn_mse)
y_predKnn = y_predKnn.reshape(360,360)
```

### <font color='blue'> Lasso Regressor

```python
from sklearn.linear_model import Lasso
lsregressor = Lasso(max_iter=100, random_state=42)

# Train the model
lsregressor.fit(X_train, y_train.values.ravel())
```

```python
# Test the model
y_predLs = lsregressor.predict(X_test)
Ls_mse = mean_squared_error(y_test,y_predLs)
print("Mean Squared Error :", Ls_mse)
y_predLs = y_predLs.reshape(360,360)
```

### <font color='blue'> Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor
# Model
gbregressor = GradientBoostingRegressor(n_estimators=100, random_state=42)

# Train the model
gbregressor.fit(X_train, y_train.values.ravel())
```

```python
# Test the model
y_predGb = gbregressor.predict(X_test)
Gb_mse = mean_squared_error(y_test,y_predGb)
print("Mean Squared Error :", Gb_mse)
y_predGb = y_predGb.reshape(360,360)
```

### <font color='blue'> Multi Layer Perceptron Regressor

```python
from sklearn.neural_network import MLPRegressor
#Model
mlpregressor = MLPRegressor(hidden_layer_sizes=(100,),random_state=42)

# Train the model
mlpregressor.fit(X_train, y_train.values.ravel())
```

```python
# Test the model
y_predMlp = mlpregressor.predict(X_test)
Mlp_mse = mean_squared_error(y_test,y_predMlp)
print("Mean Squared Error :", Mlp_mse)
y_predMlp = y_predMlp.reshape(360,360)
```

### <font color='blue'> Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
# Model
dtregressor = DecisionTreeRegressor(random_state=42)

# Train the model
dtregressor.fit(X_train, y_train.values.ravel())
```

```python
# Test the model
y_predDt = dtregressor.predict(X_test)
Dt_mse = mean_squared_error(y_test,y_predDt)
print("Mean Squared Error :", Dt_mse)
y_predDt = y_predDt.reshape(360,360)
```

### <font color='blue'> Ridge Regressor

```python
from sklearn.linear_model import Ridge
# Model
rdregressor = Ridge(random_state=42)

# Train the model
rdregressor.fit(X_train, y_train.values.ravel())
```

```python
# Test the model
y_predRd = rdregressor.predict(X_test)
Rd_mse = mean_squared_error(y_test,y_predRd)
print("Mean Squared Error :", Rd_mse)
y_predRd = y_predRd.reshape(360,360)
```
